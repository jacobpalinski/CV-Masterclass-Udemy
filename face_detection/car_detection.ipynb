```python
# Import OpenCV
import cv2
import matplotlib.pyplot as plt
```

```python
# Import image
image = cv2.imread('car.jpg')
```

```python
# Create classifier
car_detector = cv2.CascadeClassifier(r'cars.xml') 
```

```python
# Detect vehicles on greyscale image
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detections = car_detector.detectMultiScale(image_gray, scaleFactor=1.07, minNeighbors=2, minSize=(1,1), maxSize=(73,73))
for (x, y, w, h) in detections:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
